```python
# Importing basic libraries
import os
import random
import shutil
from matplotlib import pyplot
from matplotlib.image import imread
%matplotlib inline
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import Convolution2D as Conv2D
from tensorflow.keras.layers import Flatten
import tensorflow as tf
from keras.layers import Conv2D, MaxPool2D, Flatten, Dense, InputLayer, BatchNormalization, Dropout
import keras.layers as layers
```

```python
#Pre-processed data with Canny filter
training_path = '/content/drive/MyDrive/Our_Data/Thesis/LargeData/CannyData/train'
testing_path = '/content/drive/MyDrive/Our_Data/Thesis/LargeData/CannyData/test'
```

```python
#Data pre-processed with the cropping method
training_path = '/content/drive/MyDrive/Our_Data/Thesis/LargeData/ScroppedData/train'
testing_path = '/content/drive/MyDrive/Our_Data/Thesis/LargeData/ScroppedData/test'
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
BatchSize = 32
xy_shape = 64
n_epochs = 20
```

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(xy_shape, xy_shape, 3)))
model.add(layers.AveragePooling2D())
model.add(layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'))
model.add(layers.AveragePooling2D())
model.add(Flatten())

model.add(Dense(400, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(250, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(125, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(64, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(32, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
dataGen = ImageDataGenerator(rescale=1.0/255.0)
train_gen = dataGen.flow_from_directory(training_path,class_mode="binary",target_size=(xy_shape, xy_shape),
                                         batch_size=BatchSize)
test_gen = dataGen.flow_from_directory(testing_path,class_mode="binary",target_size=(xy_shape, xy_shape),
                                        batch_size=BatchSize)
class_labels = test_gen.class_indices
class_names = {value:key for (key, value) in class_labels.items()}
```

```python
#Trainng using data pre-processed with the Canny filter
history2 = model.fit(train_gen, steps_per_epoch=len(train_gen),validation_data=test_gen, validation_steps=len(test_gen),
                              epochs=n_epochs, verbose=1, use_multiprocessing=True)
```

```python
#Trainng using data pre-processed with the cropping method
history2 = model.fit(train_gen, steps_per_epoch=len(train_gen),validation_data=test_gen, validation_steps=len(test_gen),
                              epochs=n_epochs, verbose=1, use_multiprocessing=True)
```

```python
from tensorflow.keras.preprocessing.image import img_to_array
from tensorflow.keras.preprocessing.image import load_img
import os
def load_image(filename, xy_shape=xy_shape):
    # load the image
    img = load_img(filename, target_size=(xy_shape, xy_shape))
    # convert to array
    img = img_to_array(img)
    # reshape into a single sample with 3 channels
    img = img.reshape(1, xy_shape, xy_shape, 3)
    img = img.astype('float32')/255
    return img
```
